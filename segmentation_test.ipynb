# Task 2 - Evaluation

```python
from utils import semisupervised_segmentation
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_names = ["2007_000121", "2007_000123", "2007_000063"]
voc_ext = ".jpg"
gt_ext = ".png"

voc_dir = "data/voc12/"
gt_dir = "data/gt/"

images = []
gt_images = []

images_gray = []

for name in image_names:
    img = cv2.imread(voc_dir + name + voc_ext)
    gt = cv2.imread(gt_dir + name + gt_ext)
   
    images.append(img)
    gt_images.append(gt)

    images_gray.append(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
```

## Image 1

```python
segmented, final_mask, created_mask = semisupervised_segmentation(images[0])
image_gt_gray = cv2.cvtColor(gt_images[0], cv2.COLOR_BGR2GRAY)
image_gt_gray = np.where(image_gt_gray > 0, 1, 0)
```

```python
fig = plt.figure(figsize=(10, 8))

rows = 2
cols = 2

fig.add_subplot(rows, cols, 1)
plt.axis("off")
plt.title("GrabCut Input Mask")
plt.imshow(created_mask, cmap="gray")

fig.add_subplot(rows, cols, 2)
plt.axis("off")
plt.title("Final Mask")
plt.imshow(final_mask, cmap="gray")

fig.add_subplot(rows, cols, 3)
plt.axis("off")
plt.title("Ground Truth Mask")
plt.imshow(image_gt_gray, cmap="gray")

fig.add_subplot(rows, cols, 4)
plt.axis("off")
plt.title("Segmented with Our Mask")
plt.imshow(segmented[:,:,::-1])
```

```python
dice = np.sum(final_mask[image_gt_gray == 1])*2.0 / (np.sum(final_mask) + np.sum(image_gt_gray))
print(dice)
```
